```python
import itertools
```

```python
hex_values = [hex(x)[2:] for x in range(16)]
```

```python
hex_ = list(itertools.product(hex_values, hex_values))
```

```python
hex_ = [x[0] + x[1] for x in hex_]
```

```python
text = [x for x in text if x in hex_]
```
